# A Visualization demonstrating impact of Lockdown Strengency on COVID infection rate and death toll

### Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

### Read COVID cases and death dataset, filter and subset as well

```python
covid_data=pd.read_excel('../visualisation proj/data/COVID-19-geographic-disbtribution-worldwide.xlsx')
# covid_data= covid_data[covid_data['countryterritoryCode'].isin(['IRN','FRA','ITA', 'DEU','ESP'])]
covid_data = covid_data[['countryterritoryCode', 'dateRep', 'cases', 'deaths']]
covid_data.dateRep=pd.to_datetime(covid_data.dateRep, format='%Y-%m-%d', errors='coerce')
covid_data.head(5)
```

### Read COVID lockdown world data, filter and subset it

```python
lockdown_data=pd.read_csv('../visualisation proj/data/covid-lockdown-world.csv',dtype={'RegionCode':'string', 'RegionName':'string'})
# lockdown_data = lockdown_data[lockdown_data['CountryName'].isin(["Iran", "France", "Italy", "Spain", "Germany"])]
lockdown_data = lockdown_data[lockdown_data['Jurisdiction'] == 'NAT_TOTAL']
lockdown_data = lockdown_data[['CountryName','CountryCode','Date', 'StringencyIndex', 'StringencyIndexForDisplay','C1_School closing', 'C2_Workplace closing', 'C3_Cancel public events', 'C4_Restrictions on gatherings', 'C5_Close public transport', 'C6_Stay at home requirements', 'C7_Restrictions on internal movement', 'C8_International travel controls']]
lockdown_data.Date=pd.to_datetime(lockdown_data.Date, format='%Y%m%d', errors='coerce')

lockdown_data.head(5)
```

### Merge two datasets to get all data we need in one place

```python
new_df = pd.merge(covid_data, lockdown_data,  how='inner', left_on=['countryterritoryCode','dateRep'], right_on = ['CountryCode','Date'])

new_df
```

# drop extra culumns

```python
new_df.drop(['dateRep'], axis=1,inplace=True)
```

# Save new_dataset to csv

```python
new_df.to_csv('../datasets/covid-strengency-impact2.csv')
```

# select 5 countries

```python
df_ir=new_df.loc[new_df['CountryName']=="Iran",:]
df_france=new_df.loc[new_df['CountryName']=="France",:]
df_italy=new_df.loc[new_df['CountryName']=="Italy",:]
df_spain=new_df.loc[new_df['CountryName']=="Spain",:]
df_germany=new_df.loc[new_df['CountryName']=="Germany",:]
selected=[df_ir,df_france,df_italy,df_spain,df_germany]
```

# fill missing values

```python
for country in selected:
    country.cases.fillna(method='bfill',inplace=True)
    country.deaths.fillna(method='bfill',inplace=True)
```

```python
for country in selected:
    country.StringencyIndex.fillna(method='bfill',inplace=True)
```

# Detecting outlayers

```python
for country in selected:
    country['expected_cases']=(country['cases'].shift(1)+country['cases'].shift(-1))/2
```

```python
for country in selected:
    isoutlier=np.abs((country.loc[:,'expected_cases']-country.loc[:,'cases'])/country.loc[:,'expected_cases'])>0.8
    country.loc[isoutlier,'cases']=country.loc[isoutlier,'expected_cases']
```

```python
min_start=min(ds.start)
max_end=max(ds.end)
```

```python
df_mean=new_df.loc[(new_df['Date']>=min_start) & (new_df['Date']<=max_end)].groupby(by=['CountryName']).mean().reset_index()
```

```python
meanCases=dict()
for country in ds.Country:
    if (len(df_mean.loc[(df_mean['CountryName']==country)].cases.values)!=0):
            meanCases[country]=df_mean.loc[(df_mean['CountryName']==country)].cases.values[0]

meanCases
```

```python
meanDeaths=dict()
for country in ds.Country:
    if (len(df_mean.loc[(df_mean['CountryName']==country)].deaths.values)!=0):
            meanDeaths[country]=df_mean.loc[(df_mean['CountryName']==country)].deaths.values[0]

meanDeaths
```

```python
for country in meanCases.keys():
    ds.loc[(ds['Country']==country),'mean_cases']=(meanCases[country])
```

```python
for country in meanDeaths.keys():
    ds.loc[(ds['Country']==country),'mean_deaths']=(meanDeaths[country])
```
