```python
import seaborn as sns
```

```python

df_ir_corr=pd.concat([df_ir.loc[:,'C1_School closing':'C8_International travel controls'],df_ir.loc[:,'Cases':'ConfirmedDeaths']],axis=1,join='inner').drop(['C1_Flag','C2_Flag','C3_Flag','C4_Flag','C5_Flag','C6_Flag','C7_Flag','ConfirmedDeaths','ConfirmedCases'],axis=1)
df_ir_corr.insert(10,'CasesDeathRate',df_ir_corr.Deaths/df_ir_corr.Cases)
df_ir_corr.insert(11,'deathsPctChange',df_ir_corr.Deaths.pct_change(periods=14))
df_ir_corr.insert(12,'casesPctChange',df_ir_corr.Cases.pct_change(periods=14))
df_ir_corr.insert(13,'ratePctChange',df_ir_corr.CasesDeathRate.pct_change(periods=14))
```

```python
df_ir_corr.fillna(0)
```

```python
df_ir_corr.info
```

```python
df_ir['C4_Restrictions on gatherings'].value_counts()
```

```python
df_ir_corr['C1_School closing'].corr(df_ir_corr.CasesDeathRate)
```

```python
correlation=df_ir_corr.corr()
sns.heatmap(correlation,annot=True)
```

```python
ir_corr_dic8={'constraint':[],'corr_with_deathsPctChange':[],'corr_with_casesPctChange':[]}
for cl in df_ir_corr.columns[:8]:
    ir_corr_dic8['constraint'].append(cl)
    ir_corr_dic8['corr_with_deathsPctChange'].append(df_ir_corr[cl].corr(df_ir_corr.deathsPctChange))
    ir_corr_dic8['corr_with_casesPctChange'].append(df_ir_corr[cl].corr(df_ir_corr.casesPctChange))
```

```python
pd.DataFrame(ir_corr_dic8)
```

```python
ir_corr_dic={'constraint':[],'corr_with_deaths':[],'corr_with_Cases':[]}
for cl in df_ir_corr.columns[:8]:
    ir_corr_dic['constraint'].append(cl)
    ir_corr_dic['corr_with_deaths'].append(df_ir_corr[cl].corr(df_ir_corr.Deaths))
    ir_corr_dic['corr_with_Cases'].append(df_ir_corr[cl].corr(df_ir_corr.Cases))
    
   
    
```

```python
pd.DataFrame(ir_corr_dic)
```

```python

    
constraint=[]
corr_with_CasesDeathRate=[]
for cl in df_ir_corr.columns[:8]:
    constraint.append(cl)
    corr_with_CasesDeathRate.append(df_ir_corr[cl].corr(df_ir_corr.CasesDeathRate))
    
crr=pd.DataFrame(zip(constraint,corr_with_CasesDeathRate),columns=['constraint','corr_with_CasesDeathRate'] )
crr.sort_values(by=['corr_with_CasesDeathRate'])
```
