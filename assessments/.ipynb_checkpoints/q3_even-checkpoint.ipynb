```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('income.csv')
```

```python
work_hours=df['hours-per-week']
age=df['age']
plt.figure(figsize=(8,8))
plt.title("scatter plot work_hours based on age ")
plt.ylabel('work_hours')
plt.xlabel('age')

plt.scatter( age,work_hours, marker='.',alpha=0.2,color='blue')
iran_df=df.loc[df['native-country']=='Iran']
iran_work_hours=iran_df['hours-per-week']
iran_age=iran_df['age']
plt.scatter( iran_age,iran_work_hours, marker='o',alpha=0.5,color='red')
plt.legend(['world','Iran'])
```

```python
countries=df['native-country'].unique()
black_rate=[]
for country in countries:
    black_rate.append((country,(len(df.loc[(df['native-country'] ==country) & (df['race'] =='Black')])/len(df.loc[(df['native-country'] ==country)]))*100))
black_rate=sorted(black_rate, key = lambda x: x[1] ,reverse=True)  

countries=[k for k,v in black_rate[:5]]
rate=[v for k,v in black_rate[:5]]

fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.bar(countries,rate)
plt.xlabel('country')
plt.ylabel('rate of black race')
plt.title('rate of black race in 5 highest rate countries ')
plt.grid()
plt.show()
```
