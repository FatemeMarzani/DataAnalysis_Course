```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('income.csv')
df
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
incom_male = ['<=50k', '>50k']
number = [len(df.loc[(df['income'] =='<=50K') & (df['gender']  == 'Male')]),len(df.loc[(df['income'] =='>50K') & (df['gender']  == 'Male')])]
ax.bar(incom_male,number)
Female = ['<=50k', '>50k']
number = [len(df.loc[(df['income'] =='<=50K') & (df['gender']  == 'Female')]),len(df.loc[(df['income'] =='>50K') & (df['gender']  == 'Female')])]
ax.bar(incom_male,number)
plt.legend(['Male','Female'])
plt.xlabel('income category')
plt.ylabel('number of person')
plt.title('number of person based on income ')
plt.grid()
plt.show()
```

```python
gender_column =df['gender']  == 'Female'
df.groupby(["marital-status", gender_column]).size().unstack(level=1).plot(kind='bar')
plt.legend(['Female','Male'])
plt.xlabel('marital-status')
plt.ylabel('number of person')
plt.title('number of person based on marital-status ')
plt.grid()
plt.show()
```
