```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
oil=pd.read_csv('../datasets/oil_consumption_per_cap.csv',index_col="country")
print(oil.columns)
oil.columns=[int(s) for s in oil.columns]
oil=oil[range(1971,2015)]
oil=oil.fillna(method='bfill',axis=1)
display(oil)
print(oil.index)
print(oil.columns)
```

```python
electricity=pd.read_csv('../datasets/electricity_use_per_person.csv',index_col="country")
electricity.columns=[int(s) for s in electricity.columns]
electricity=electricity.loc[oil.index,:]
electricity=electricity[range(1971,2015)]
electricity=electricity.fillna(method='bfill',axis=1)
display(electricity)
print(electricity.index)
print(electricity.columns)

```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
plt.plot(electricity.loc['Iran',:])
plt.subplot(2,1,2)
plt.plot(oil.loc['Iran',:])
```

```python
plt.subplot(1,2,1)
plt.imshow(oil)
plt.subplot(1,2,2)
plt.imshow(electricity)
```

```python
electricity['mean_usage']=electricity.mean(axis=1)
print(electricity.columns)
# electricity=electricity.sort_values(by='mean_usage',ascending=False)
electricity=electricity.drop(['mean_usage'],axis=1)
print(electricity.columns)
# electricity
```

```python
plt.plot(electricity.mean(axis=0))
plt.plot(electricity.loc['Iran',:],'r--')
plt.legend(['World','Iran'])
plt.xlabel('Year')
plt.ylabel('electricity connsumption per capita (KWh/year)')
plt.title('electricity connsumption, Iran vs world')
plt.grid()
```

```python
plt.plot(oil.mean(axis=0))
plt.plot(oil.loc['Iran',:],'r--')
plt.legend(['World','Iran'])
plt.xlabel('Year')
plt.ylabel('oil connsumption per capita (BL/year)')
plt.title('oil connsumption, Iran vs world')
plt.grid()
```

```python
plt.figure(figsize=(6,6))
plt.subplot(2,1,1)
plt.hist(electricity.loc[:,2014],bins=np.arange(0,25000,1000),alpha=0.5);
plt.legend(['2014'])
plt.ylim((0,12))
plt.ylabel('Count')

plt.subplot(2,1,2)
plt.hist(electricity.loc[:,1994],bins=np.arange(0,25000,1000),alpha=0.5);
plt.legend(['1994'])
plt.ylim((0,12))
plt.ylabel('Count')
plt.ylabel('Kwh/year/person')
```

```python
plt.figure(figsize=(10,10))
for country in electricity.index:
    plt.plot(electricity.loc[country,:],oil.loc[country,:],'.',color='gray',alpha=0.2)
country='Iran'
plt.plot(electricity.loc[country,:],oil.loc[country,:],'.',color='red',alpha=1)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Annual electricity connsumption  KWh per capita')
plt.ylabel('Annual oil connsumption BBl per capita')
plt.title('annual energy connsumption, electricity vs oil across countries and Iran')
plt.grid()
```
