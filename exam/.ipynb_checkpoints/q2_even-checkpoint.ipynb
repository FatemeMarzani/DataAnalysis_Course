```python
import pandas as pd
import numpy as np
from functools import reduce
def female_worker_percent(series): 
    return (len(list(filter(lambda x: x == 'Female', series))) / len(series) * 100)

income = pd.read_csv('income.csv')
income['count'] = 0
x = income.groupby('native-country').agg({ 'count': len, 'hours-per-week': 'mean', 'gender': female_worker_percent})
print(x)
```
